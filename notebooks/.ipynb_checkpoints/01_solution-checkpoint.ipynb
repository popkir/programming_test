```python
def get_daily_prices_yahoo(symbol: str) -> pd.DataFrame:
    """
    Fetches full history of daily stock prices from Yahoo Finance.
    
    Args:
        symbol (str): stock symbol to fetch
        
    Returns:
        pandas.DataFrame: Daily ticker data
    """
    ticker = yf.Ticker(symbol)
    data = ticker.history(period='max').reset_index()
    data.columns = [c.lower() for c in data.columns]
    return data
```

```python
def save_daily_prices(df: pd.DataFrame, symbol: str) -> None:
    """
    Saves daily price data for a symbol as csv.
    
    Args:
        df (pandas.DataFrame): data to save
        symbol (str): stock symbol used to create file name
    """
    root = '../data/01_daily_prices'
    filename = symbol + '.csv'
    path = os.path.join(root, filename)
    df.to_csv(path, index=False)

def load_daily_prices(symbol: str) -> pd.DataFrame:
    """
    Loads daily price data for a symbol from csv.
    Sets date formats.
    Assumes there is 'date' columns with timestamp, and the rest of columns are numerical.

    Args:
        symbol (str): stock symbol to load

    Returns:
        pandas.DataFrame: daily price data
    """
    root = '../data/01_daily_prices'
    filename = symbol + '.csv'
    path = os.path.join(root, filename)
    df = pd.read_csv(path)
    
    # Set data formats
    df = df.set_index('date')
    df.index = pd.to_datetime(df.index)
    df = df.astype(float)
    df = df.reset_index()
    
    return df
```

### Question 2
- Compute returns $$r_i = (P_{i+1} - P_{i}) / P_{i},$$ where $i$ is today, and $i+1$ is tomorrow
- So, $r_i$ represents the return of an investment buying / selling today (day $i$) to selling / buying tomorrow (day $i+1$).
- Can you compute some statistics on the return time series?
- For each stock
     - What is its distribution? Gaussian, Student-t with $\nu$ degrees of freedom?
     - Is the distribution stationary over time? (Hint: look for stationarity tests)
     - Is there any auto-correlation in the time series?
     - Are there any interesting relationships between the returns of different stocks at the same time (contemporaneous) or different times? (Hint: look at correlation, Granger causality, PCA decomposition of returns)

```python
def compute_return(ts: pd.Series) -> pd.Series:
    """
    Computes the return timeseries from price timeseries.
    Assumes that dates are sorted in ascending order, and prices are positive.
    
    Args:
        ts (pd.Series): original timeseries (e.g. stock price)
    
    Returns:
        pd.Series: timeseries of returns
    """
    assert all(ts.dropna()>0)
    return (ts.shift(-1) - ts)/ts
```

```python
from scipy.stats import ks_1samp, norm, kstest
from scipy.stats import t as student_t
```

```python
pvalue_threshold = 0.01
min_dof = 0
max_dof = 30
```

```python
student_pvalues = pd.DataFrame(columns=symbols)
```

```python
for s in symbols:
    ts = returns[s].dropna()
    
    print('\n' + s)
    
    for dof in range(min_dof, max_dof):
        fit_params = student_t.fit(ts, fdf=dof)
        cdf = student_t(*fit_params).cdf
        test_results = kstest(ts, cdf=cdf)#
        
        student_pvalues.loc[dof, s] = test_results.pvalue
        print(test_results)
        
        if test_results.pvalue >= pvalue_threshold:
            print(f'Can not reject that {s} returns are t-distributed with {dof} degrees of freedom')
        else:
            print(f'Can reject that {s} returns are t-distributed with {dof} degrees of freedom')
```

#### Granger causality test

Very curious results. At 99% significance, with up to 150 days included in the regressors:
- GOOG returns can be used to predict AAPL returns within a week
- AAPL returns are marginally useful to predict GOOG returns 149 days ahead
- FB returns can be used to predict AAPL returns
- AAPL returns are not useful to predict future FB returns
- FB returns can be used to predict GOOG returns
- GOOG returns can be used to predict FB returns

```python
from statsmodels.tsa.stattools import grangercausalitytests
from itertools import combinations
```

```python
maxlag = 150
pvalue_threshold = 0.05
```

```python
granger_p_values = pd.DataFrame()
```

```python
for s1, s2 in combinations(symbols, 2):
    print('\n' + s2 + '->' + s1)
    x = returns[[s1, s2]].dropna().values
    s2_affects_s1 = grangercausalitytests(x, maxlag, verbose=False)
    for lag in s2_affects_s1.keys():
        tests = s2_affects_s1[lag][0]
        p_values = [v[1] for v in tests.values()]
        if any([p>=pvalue_threshold for p in p_values]):
            print(f"Lag {lag}: Reject non-causality by at least one test;\np-values {p_values}")

        granger_p_values.loc[lag, s2 + '->' + s1] = np.mean(p_values)
        
    print('\n' + s1 + '->' + s2)
    x = returns[[s2, s1]].dropna().values
    s1_affects_s2 = grangercausalitytests(x, maxlag, verbose=False)
    for lag in s1_affects_s2.keys():
        tests = s1_affects_s2[lag][0]
        p_values = [v[1] for v in tests.values()]
        if any([p>=pvalue_threshold for p in p_values]):
            print(f"Lag {lag}: Reject non-causality by at least one test;\np-values {p_values}")
        granger_p_values.loc[lag, s1 + '->' + s2] = np.mean(p_values)
    
```

```python
for c in granger_p_values.columns:
    fig, ax = plt.subplots(1, 1, figsize=(16,9))
    ax.set_title(f'p-values of Granger causality test for {c}')
    ax.set_ylabel('Mean p-value of 4 tests')
    ax.set_xlabel('Lag, trading days')
    ax.plot(granger_p_values.index, granger_p_values[c], label=f'p-value for {c}')
    ax.hlines(0.01, xmin=0, xmax=len(granger_p_values)+1, label='1% significance level', color='C1')
    ax.grid()
    ax.legend()
    fig.savefig(f'../visualizations/06_granger_causality/{c}.png')
```

### Question 4
- For a PNL time series, define the Sharpe Ratio to be $$SR = E[pnl_i] / \sqrt{Var[pnl_i]}$$
- Assume that $f_i$ has a given hit rate $= 1/n \sum_i (sign(f_i) == sign(r_i))$
    - Hit rate is a measure of accuracy and represents how often we predict the direction of the stock movement correctly
- Remember that $f_i$ is a random variable, and so you can draw different realizations of it
- What is the distribution of the $SR(pnl_i)$ given $f_i$?
    - Ideally, you want to generate a plot varying hit_rate from 0 to 1.0 on the x-axis and computing the corresponding distribution of SR on the y-axis
    - Note that $pnl_i$ is a random var since it is a function of the random variable f_i
    - You can plot the distribution as a box plot or using just mean and std err

```python
def make_strategy(returns_df: pd.DataFrame,
                  symbol: str,
                  hit_rate: float) -> pd.Series:
    """
    Make strategy with given hit rate w.r.t. given returns of a given ticker.
    Absolute values are sampled from std normal.
    
    Args:
        returns_df (pd.DataFrame): dataframe with returns data, has 'date' & symbol ticker columns
        symbol(str): ticker symbol, string
        hit_rate(float): a number in [0,1] that defines the hit rate of the strategy
        
    Returns:
        pd.Series: actions according to the strategy
    """
    df = returns_df[['date', symbol]].dropna().rename(columns={symbol: 'return'})
    
    # Sample from uniform on [0,1)
    samples = np.random.rand(len(df))
    # Figure which signs should match
    should_match_signs = 2 * (samples <= hit_rate) - 1
    # Make signs
    action_signs = should_match_signs * np.sign(df['return'])
    # Make absolute values
    magnitudes = np.abs(norm.rvs(size=len(df)))
    
    strategy = magnitudes * action_signs
    
    return strategy
```

```python
def compute_hit_rate(ts1: pd.Series, ts2: pd.Series) -> float:
    """
    Compute & return the hit rate between two series, i.e. in what fraction of entries do the signs match.
    Series should have same length!
    
    Args:
        ts1 (pd.Series): the first series
        ts2 (pd.Series): the second series

    Returns:
        float: the hit rate
    """
    hit = np.sign(ts1)==np.sign(ts2)
    hit_rate = hit.astype(int).mean()
    return hit_rate
```

```python
def sharpe_ratio(ts: pd.DataFrame) -> pd.Series:
    """
    Compute & return the Sharpe ratio from mutiple samples of timeseries.
    
    Args:
        ts (pd.DataFrame): dataframe where each column is a timeseries

    Returns:
        pd.Series: series with Sharpe ratios at each timestep
    """
    sharpe = ts.mean(axis=1) / ts.std(axis=1)
    return sharpe
```

### Question 5
- What if, instead of hit rate, you use root mean square error (RMSE) between $f_i$ and $r_i$?
- How does the distribution of SR change?
- What if you use mean absolute deviation (MAD) of $f_i$ and $r_i$?
- Make the code flexible to changing input measures (hit rate, RMSE, MAD) and output measures

```python
def make_strategy_rmse(returns_df: pd.DataFrame,
                       symbol: str,
                       rmse: float) -> pd.Series:
    """
    Make strategy with given RMSE w.r.t. given returns of a given ticker.
    Made by computing random strategy orthogonal to returns and adding it to returns
    
    Args:
        returns_df (pd.DataFrame): dataframe with returns data, has 'date' & symbol ticker columns
        symbol(str): ticker symbol, string
        rmse(float): a number >= 0 that defines the RMSE of the strategy vs returns timeseries
        
    Returns:
        pd.Series: actions according to the strategy
    """
    df = returns_df[['date', symbol]].dropna().rename(columns={symbol: 'return'})
    returns = df['return']

    # Sample from normal
    vector = np.random.normal(size=len(df))
    # Get orthogonal part to the returns
    orthogonal = vector - returns * np.dot(vector, returns) / np.linalg.norm(returns)**2
    # Get unit orthogonal vector
    unit_orthogonal = orthogonal / np.linalg.norm(orthogonal)
    
    
    # Scale by RMSE
    diff = unit_orthogonal * rmse * np.sqrt(len(df))
    
    strategy = returns + diff
    
    return strategy
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 9))
ax.set_title(f'Sharpe ratio distributions for strategies on {symbol}')
sns.boxplot(x="rmse", y="sharpe", data=sharpe_master_df, ax=ax)
ax.grid()
fig.savefig(f'../visualizations/09_sharpe_vs_rmse/{symbol}.png')
```
